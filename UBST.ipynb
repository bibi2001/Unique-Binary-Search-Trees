```python
import random
#exercicio 1a

def randomList(size):
	list = []
	for x in range(size-1):
		list.append(random.randint(0, 1000000))
	list.sort()
	return list
print(randomList(100))
```

```python
import numpy as np
#exercicio 1b

def randomNumpy(size):
	arr = np.zeros(size)
	for x in range(size-1):
		arr[x] = random.randint(0, 1000000)
	arr.sort()
	return arr
print(randomNumpy(100))
```

```python
#exercicio 2a

def search_in(x1, list1):
     return x1 in list1
print(search_in(0,list1 = [1,1]))
```

```python
#exercicio 2b

def search_pes(x2, list2):
    for i in list2:
        if i==x2:
            return True
    return False
print(search_pes(1,list2 = [1,1]))
```

```python
#exercicio 2c aux

def binarySearch(list3, left, right, x3):
	if right >= left:
		mid = left + (right - left) // 2
		if list3[mid] == x3:
			return mid #se encontrar
		elif list3[mid] > x3:
			return binarySearch(list3, left, mid-1, x3) #vai para a metade esquerda
		else:
			return binarySearch(list3, mid + 1, right, x3) #vai para a metade direita
	else:
		return -1 #se não encontrar
```

```python
#exercicio 2c
def  search_bin(x3, list3):
	return binarySearch(list3, 0, len(list3)-1, x3) !=-1
print(search_bin(2,list3 = [1,1]))
```

```python
#exercicio 4

listINi = []
listSEQi = []
listBSii = []
listSEQii = []
listINii = []
listBSi = []
arrINi = []
arrSEQi = []
arrBSi =[]
arrINii=[]
arrSEQii=[]
arrBSii=[]
runList = range(1000, 100000, 1000)

for number in runList:
	listRand = randomList(number)
	arrRand = randomNumpy(number)
	element = random.randint(1,number)

	listINi.append(sum(repeat(lambda :search_in(1000001,listRand), number=1, repeat=36)) /36)
	listSEQi.append(sum(repeat(lambda :search_pes(1000001,listRand), number=1, repeat=36)) /36)
	listBSi.append(sum(repeat(lambda :search_bin(1000001,listRand), number=1, repeat=36)) /36)

	listINii.append(sum(repeat(lambda :search_in(listRand[element],listRand), number=1, repeat=36)) /36)
	listSEQii.append(sum(repeat(lambda :search_pes(listRand[element],listRand), number=1, repeat=36)) /36)
	listBSii.append(sum(repeat(lambda :search_bin(listRand[element],listRand), number=1, repeat=36)) /36)

	arrINi.append(sum(repeat(lambda :search_in(1000001,arrRand), number=1, repeat=36)) /36)
	arrSEQi.append(sum(repeat(lambda :search_pes(1000001,arrRand), number=1, repeat=36)) /36)
	arrBSi.append(sum(repeat(lambda :search_bin(1000001,arrRand), number=1, repeat=36)) /36)

	arrINii.append(sum(repeat(lambda :search_in(arrRand[element],arrRand), number=1, repeat=36)) /36)
	arrSEQii.append(sum(repeat(lambda :search_pes(arrRand[element],arrRand), number=1, repeat=36)) /36)
	arrBSii.append(sum(repeat(lambda :search_bin(arrRand[element],arrRand), number=1, repeat=36)) /36)
```

```python
import matplotlib.pyplot as pltListi
pltListi.plot(runList, listINi, label = "In list i.") #azul
pltListi.plot(runList, listSEQi, label = "SEQ list i.") #laranja
pltListi.plot(runList, listBSi, label = "BS list i.") #verde
pltListi.legend()
```

```python
#exercicio 6

rangeINi = []
rangeINii = []
rangeSEQi = []
rangeSEQii =[]
rangeBSi = []
rangeBSii = []

for number in runList:
	runList1 = range(number)
	element = random.randint(1,number)

	rangeINi.append(sum(repeat(lambda :search_in(runList1[element],runList1), number=1, repeat=36)) /36)
	rangeINii.append(sum(repeat(lambda :search_in(1000001,runList1), number=1, repeat=36)) /36)

	rangeSEQi.append(sum(repeat(lambda :search_pes(runList1[element],runList1), number=1, repeat=36)) /36)
	rangeSEQii.append(sum(repeat(lambda :search_pes(1000001,runList1), number=1, repeat=36)) /36)

	rangeBSi.append(sum(repeat(lambda :search_bin(runList1[element],runList1), number=1, repeat=36)) /36)
	rangeBSii.append(sum(repeat(lambda :search_bin(1000001,runList1), number=1, repeat=36)) /36)
```

```python
import matplotlib.pyplot as plotIn
plotIn.plot(runList, rangeINi, label = "In i.") #azul
plotIn.plot(runList, rangeINii, label = "In ii.") #laranja
plotIn.legend()
```

```python
import matplotlib.pyplot as plotPes
plotPes.plot(runList, rangeSEQi, label = "Seq i.") #azul
plotPes.plot(runList, rangeSEQii, label = "Seq ii.") #laranja
plotPes.legend()
```

```python
import matplotlib.pyplot as plotBS
plotBS.plot(runList, rangeBSi, label = "BS i.") #azul
plotBS.plot(runList, rangeBSii, label = "BS ii.") #laranja
plotBS.legend()
```
